```python
import pandas as pd
```

```python
data=pd.read_excel('/content/Book1.xlsx')
```

```python
data1=pd.read_csv('/content/Data Ops - TPR Client Table QA_GDS Test copy TPR Performance Table_Table (63).csv')
```

```python
# split the tbl column
data[['temp','channel','brand','site']] = data['tbl'].str.split('_', expand=True)

# temp column remove
data = data.drop(columns=['temp'])

# check output
print(data[['tbl','channel','brand','site']].head())
```

```python
data.columns
```

```python
perf= data[['date','brand','channel','site',
             'spend_usd_engg',
             'demand_usd_engg',
             'clicks_engg',
             'sessions_engg',
             'impressions_engg']]
```

```python
perf['brand'] = perf['brand'].replace({
    'COH': 'Coach',
    'KS': 'Kate Spade'
})
```

```python
# Channel changes
perf['channel'] = perf['channel'].replace({
    'Search': 'Paid Search',
    'Social': 'Paid Social'
})

# Site changes based on brand + site
perf.loc[
    (perf['site'] == 'RTL') &
    (perf['brand'] == 'Coach'),
    'site'
] = 'Coach Retail'

perf.loc[
    (perf['site'] == 'RTL') &
    (perf['brand'] == 'Kate Spade'),
    'site'
] = 'K.Spade Retail'

perf.loc[
    (perf['site'] == 'OTL') &
    (perf['brand'] == 'Coach'),
    'site'
] = 'Coach Outlet'

perf.loc[
    (perf['site'] == 'OTL') &
    (perf['brand'] == 'Kate Spade'),
    'site'
] = 'K.Spade Outlet'

# Check output
print(perf[['brand','channel','site']].head())
```

```python
daily_qa= data1[['calendar_date','brand','channel','site',
             'spend_usd',
             'demand_usd',
             'clicks',
             'sessions',
             'impressions']]
```

```python
perf.columns=perf.columns.str.strip().str.lower()

daily_qa.columns=daily_qa.columns.str.strip().str.lower()

perf["date"]=pd.to_datetime(perf["date"])

daily_qa["calendar_date"]=pd.to_datetime(daily_qa["calendar_date"])
```

```python
daily_qa.head()
```

```python
# Perf aggregate
perf_grp = perf.groupby(
    ['date', 'site', 'channel', 'brand'],
    as_index=False
).agg({
    'spend_usd_engg':'sum',
    'demand_usd_engg':'sum',
    'clicks_engg':'sum',
    'sessions_engg':'sum',
    'impressions_engg':'sum'
})

# QA aggregate
daily_qa_grp = daily_qa.groupby(
    ['calendar_date', 'site', 'channel', 'brand'],
    as_index=False
).agg({
    'spend_usd':'sum',
    'demand_usd':'sum',
    'clicks':'sum',
    'sessions':'sum',
    'impressions':'sum'
})

# Merge
combined_df = pd.merge(
    perf_grp,
    daily_qa_grp,
    left_on=['date','site','channel','brand'],
    right_on=['calendar_date','site','channel','brand'],
    how='inner'
)

# Final columns
combined_df = combined_df[[
    'date',
    'site',
    'channel',
    'brand',
    'spend_usd_engg',
    'demand_usd_engg',
    'clicks_engg',
    'sessions_engg',
    'impressions_engg',
    'spend_usd',
    'demand_usd',
    'clicks',
    'sessions',
    'impressions'
]]

combined_df.head()
```

```python
# Calculate all diff %
combined_df['spend_diff_%'] = (
    abs(combined_df['spend_usd'] - combined_df['spend_usd_engg']) /
    combined_df['spend_usd_engg'].replace(0,1)
) * 100

combined_df['demand_diff_%'] = (
    abs(combined_df['demand_usd'] - combined_df['demand_usd_engg']) /
    combined_df['demand_usd_engg'].replace(0,1)
) * 100

combined_df['clicks_diff_%'] = (
    abs(combined_df['clicks'] - combined_df['clicks_engg']) /
    combined_df['clicks_engg'].replace(0,1)
) * 100

combined_df['sessions_diff_%'] = (
    abs(combined_df['sessions'] - combined_df['sessions_engg']) /
    combined_df['sessions_engg'].replace(0,1)
) * 100

combined_df['impressions_diff_%'] = (
    abs(combined_df['impressions'] - combined_df['impressions_engg']) /
    combined_df['impressions_engg'].replace(0,1)
) * 100


# Add % symbol
diff_cols = ['spend_diff_%','demand_diff_%','clicks_diff_%',
             'sessions_diff_%','impressions_diff_%']

for col in diff_cols:
    combined_df[col] = combined_df[col].round(2).astype(str) + '%'


combined_df.head()
```

```python
combined_df['date'] = pd.to_datetime(
    combined_df['date']
).dt.date
```

```python
combined_df.to_excel(
    "qa_output.xlsx",
    index=False,
    engine="openpyxl"
)
```
